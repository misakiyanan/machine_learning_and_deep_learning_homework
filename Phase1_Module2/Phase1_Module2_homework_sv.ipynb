#### 1.随机数生成六个班的考试成绩，3门考试：Python、数学、语文。每个班50人

#### 4.分别计算男女生各科成绩统计指标：最小值、最大值、平均分、中位数、标准差

```python
def course_stats(data, course_idx):
    if course_idx == 0:
        course = 'Python'
    elif course_idx == 1:
        course = '数学'
    else:
        course = '语文'
        
    min_ = np.min(data)
    max_ = np.max(data)
    mean_ = np.mean(data)
    median_ = np.median(data)
    std_ = np.std(data)
    
    print(f'{course}科目成绩：最小值为{min_}，最大值为{max_}，平均分为{mean_}，中位数为{median_}，标准差为{std_}')
    return
    
```

```python
def stats(data, sex):
    if sex == 1:
        sex_ = '女生'
    if sex == 0:
        sex_ = '男生'
    print(f'【{sex_}成绩统计】')
    
    subset = data[data[:,3]==sex]
        
    for c in range(subset.shape[1]-1):
        course_stats(subset[:,c], c)
    print()
    return
```

```python
stats(data, 0)
stats(data, 1)
```
